3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
### YOUR CODE HERE ###
#создаем 2 таблицы с количеством студентов, начавших работу над дз и сделавшими дз
done_temp = main[(main['is_homework']==True)&(pd.isna(main['finish_date'])==False)]\
                .loc[:,['course_title','module_number', 'student_id']]\
                .drop_duplicates().groupby(by=['course_title','module_number']).\
                agg(done_number=('student_id', 'count')).reset_index()
fourth_task = main[main['is_homework']==True]\
                .loc[:,['course_title','module_number','module_title', 'student_id']]\
                .drop_duplicates().groupby(by=['course_title','module_number','module_title']).\
                agg(not_done_number=('student_id', 'count')).reset_index()
#объединяем в одну таблицу и рассчитываем конверсию
fourth_task = fourth_task.merge(done_temp, on=('course_title', 'module_number'))
fourth_task['conversion'] = fourth_task['not_done_number'] / fourth_task['done_number'].shift(1)
#для первых в курсе дз конверсия посчитана неверно (точнее, она просто зачем-то посчитана)
#дальше для первых дз в курсе выставляем конверсию = 1
first_hw = fourth_task.groupby(['course_title']).agg(start_number=('module_number', 'min')).reset_index()
hw_filter = fourth_task.set_index(['course_title','module_number']).\
         index.isin(first_hw.set_index(['course_title','start_number']).index)
fourth_task.loc[hw_filter,'conversion'] = 1
fourth_task
```

```python
for i in fourth_task.groupby(by=['course_title']).groups.keys():
    result = fourth_task.groupby(['course_title']).get_group(i)
    result.plot(kind='bar',y='conversion', x='module_title',
                title=i,xlabel='Название модуля',ylabel='Конверсия',rot=90,grid=True,legend=False)
    plt.show()
```
